```python
url = "https://www.amazon.com/Disney-Mens-Donald-T-Shirt-Heather/dp/B0788YHL3H/ref=sr_1_5?c=ts&keywords=Men%27s+Novelty+T-Shirts&qid=1647094600&s=apparel&sr=1-5&ts_id=9056987011"

```

```python
# Open the driver (change the executable path to geckodriver_mac.exe or geckodriver.exe)
driver = se.webdriver.Chrome(executable_path="chromedriver")
#chrome_options = webdriver.ChromeOptions()
#chrome_options.add_argument('--proxy-server=%s' % proxies)
```

```python
driver.get(url)
```

```python
driver.page_source
```

```python
#get product title 
productTitle=driver.find_element_by_id('productTitle')
```

```python
productTitle.text
```

```python
#get product price
productPrice=driver.find_element_by_class_name('a-price-range')
```

```python
productPrice.text
```

```python
#get cout ratings 
numRatings=driver.find_element_by_id('acrCustomerReviewText')
```

```python
numRatings.text
```

```python
#get available sizes
sizemenu = driver.find_element_by_id('native_dropdown_selected_size_name')
```

```python
select = sizemenu.find_element_by_css_selector('option')
```

```python
select = sizemenu.find_elements_by_css_selector('option')
```

```python
for item in select:
    print (item.text)
```
